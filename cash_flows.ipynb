#### Шаг 2. Предобработка таблиц до экспорта в excel

```python
TransformedFintoolReferenceData = FintoolReferenceData.copy()[[
    'isincode', 'fintoolid', 'nickname', 'issuername', 'issuerinn', 'borrowername', 'borrowerinn',
    'fintooltype', 'facevalue', 'begdistdate', 'faceftname', 'nominaltype', 'coupontype', 'floatratename'
]].drop_duplicates(subset=['isincode'])\
.rename(columns={
    'isincode': "isin",
    'fintoolid': "fintool",
    'nickname': "name",
    'issuername': "issuer",
    'issuerinn': "issuer_INN",
    'borrowername': "borrower",
    'borrowerinn': "borrower_INN",
    'fintooltype': "type",
    'facevalue': "face_value",
    'begdistdate': "issue_date",
    'faceftname': "currency",
    'nominaltype': "principal_type",
    'coupontype': "coupon_type",
    'floatratename': "rate"
})
TransformedFintoolReferenceData.set_index('isin', drop=True, inplace=True)
TransformedFintoolReferenceData['issue_date'] = pd.to_datetime(TransformedFintoolReferenceData['issue_date'])
```

```python
TransformedAccruedInterestOnDate = AccruedInterestOnDate.copy().dropna(subset=["currentFaceValue"])
TransformedAccruedInterestOnDate = TransformedAccruedInterestOnDate\
.merge(FintoolReferenceData[['fintoolid', "isincode"]], how='left', left_on='fintoolId', right_on='fintoolid')\
.drop(columns=['fintoolid'])\
.rename(columns={
    "isincode": "isin", 
    'currentFaceValue': 'face_value', 
    'faceValueCurrency': 'currency'
})
TransformedAccruedInterestOnDate = TransformedAccruedInterestOnDate[['isin', 'face_value', 'currency']].set_index('isin')

#Добавление загруженных данных в результирующий датафрейм
TransformedAccruedInterestOnDate = TransformedAccruedInterestOnDate[~TransformedAccruedInterestOnDate.index.duplicated()]
TransformedFintoolReferenceData.update(TransformedAccruedInterestOnDate)
```

```python
TransformedFloaterData = FloaterData.copy()
TransformedFloaterData = pd\
.concat([TransformedFloaterData, pd.DataFrame([base[0] for base in FloaterData.bases])], axis=1)\
.drop(columns=['bases'])
TransformedFloaterData = TransformedFloaterData\
.merge(FintoolReferenceData[['fintoolid', "isincode"]], how='left', left_on='fintoolId', right_on='fintoolid')\
.drop(columns=['fintoolid'])\
.reset_index(drop=True)
TransformedFloaterData['beg_period'] = pd.to_datetime(TransformedFloaterData['beg_period'])
TransformedFloaterData['end_period'] = pd.to_datetime(TransformedFloaterData['end_period'])
TransformedFloaterData['premium_to_base'] /= 10000
```

```python
TransformedCalendar = Calendar.copy()
# TransformedCalendar.loc[TransformedCalendar['typeOperation'].isin(['O','N','P']), 'eventType'] = 'PUT'
TransformedCalendar['eventDate'] = pd.to_datetime(TransformedCalendar['eventDate'])
TransformedCalendar = TransformedCalendar.drop(columns=['typeOperation'])
TransformedCalendar = TransformedCalendar[[
    'isiNcode', 'finToolID', 'nickname', 'eventDate', 'eventType', 
    'value', 'pay1Bond', 'faceFTName', 'note', 'eventTotalVol'
]].rename(columns={
    'isiNcode': "isin",
    'finToolID': "fintool",
    'nickname': "name",
    'eventDate': "payment_date", 
    'eventType': "payment_type", 
    'value': "rate", 
    'pay1Bond': "payment_sum", 
    'faceFTName': "currency", 
    'note': "note",
    'eventTotalVol': 'event_total_vol'
})
# TransformedCalendar = TransformedCalendar[TransformedCalendar['Type'] != 'CALL']  # если нужны оферты колл, нужно аккуратно подойти к выбору typeOperation
TransformedCalendar['rate'] /= 100
TransformedCalendar.sort_values(['fintool', 'payment_date'], inplace=True)
TransformedCalendar.dropna(subset="isin", inplace=True)
TransformedCalendar.reset_index(drop=True, inplace=True)
```

#### Шаг 3. Корректировка потока

```python
bonds = TransformedCalendar\
.merge(TransformedFintoolReferenceData[["face_value", "principal_type", "coupon_type", "issue_date"]], left_on="isin", right_index=True)\
.rename(columns={"face_value": "nominal"})
bonds["nominal"] = bonds["nominal"].fillna(0)
bonds["rate"] = bonds["rate"].fillna(0)
bonds["payment_sum"] = bonds["payment_sum"].fillna(0)
bonds['event_total_vol'] = bonds['event_total_vol'].fillna(0)
bonds["name"] = bonds["name"].str.replace('[\n\t\r;,]', '.', regex=True)
bonds["note"] = bonds["note"].str.replace('[\n\t\r;,]', '.', regex=True)
```

```python
bonds.loc[bonds["payment_type"]=="DFLT", "payment_sum"] *= -1
```

```python
bonds["payment_type_NUM"] = bonds["payment_type"].map({"CPN": 1, "DIV": 1, "CONV": 1,  "CALL": 2, "MTY": 3, "DFLT": 4})
```

```python
bonds["payment_sum"] = bonds["payment_sum"].where(~(bonds['event_total_vol']==0) & ~(bonds["rate"]==0), other=0)
bonds.loc[(bonds.payment_type=="CALL")&(bonds["payment_date"]>=REPORT_DATE), "payment_sum"] = 0
```

```python
call_bonds = bonds[bonds["payment_type"].isin(["CALL", "MTY"])]
cpn_bonds = bonds[~bonds["payment_type"].isin(["CALL", "MTY"])]
bonds = pd.concat([cpn_bonds, call_bonds.groupby(["isin", "payment_date"]).last().reset_index()], axis=0).sort_values(by=["isin", "payment_date", "payment_type_NUM"]).reset_index(drop=True)
```

```python
EmptyNominal = bonds[bonds.nominal==0]["isin"]
FirstMTY = bonds[(bonds["isin"].isin(EmptyNominal))&(bonds.payment_type=="MTY")].groupby("isin").first()[["rate", "payment_sum"]]
FirstMTY["nominal"] = FirstMTY["payment_sum"]/FirstMTY["rate"]
bonds.loc[bonds.nominal==0, "nominal"] = bonds[bonds.nominal==0]["isin"].map(FirstMTY.nominal.to_dict())
```

```python
bonds['id_payment'] = bonds.groupby(['isin']).cumcount() + 1
bonds["report_date"] = pd.to_datetime(REPORT_DATE)
```

```python
bonds["MTY"] = np.where(bonds.payment_type=="MTY", bonds.payment_sum, 0)
bonds["paid_nominal"] = bonds.groupby(["isin"])["MTY"].cumsum()
bonds["remain_nominal"] = bonds["nominal"] - bonds["paid_nominal"]
```

Флоатеры

```python
floaters_isins = bonds[bonds['coupon_type'].isin(['Переменный', 'Плавающий'])].index
Floaters = bonds[bonds.index.isin(floaters_isins)]
Bonds = bonds[(bonds["payment_date"] > REPORT_DATE) & ~(bonds.index.isin(floaters_isins) | bonds["coupon_type"].isna())]
```

```python
Floaters_CPN_max = Floaters[Floaters.payment_type=="CPN"].groupby("isin").agg({"rate": "max"})
Floaters["CPN_max"] = Floaters["isin"].map(Floaters_CPN_max["rate"].to_dict())
```

```python
Future_CPN = Floaters[(Floaters["payment_type"]=="CPN") & (Floaters["payment_sum"]==0)].index
Floaters.loc[Future_CPN, "payment_sum"] = Floaters.loc[Future_CPN, "remain_nominal"]*Floaters.loc[Future_CPN, "CPN_max"]
```

```python
Floaters = Floaters[Floaters["payment_date"]>=REPORT_DATE]
```

#### Шаг 4. Объединение облигаций с постоянным и плавающим купоном в один датафрейм

```python
Bonds["is_floater"] = False
Floaters["is_floater"] = True
```

```python
bonds = pd.concat([Bonds, Floaters], axis=0).reset_index(drop=True)
```

Добавить отсутствующую оферту по каждому ISIN

```python
last_bonds = bonds.groupby("isin").last()
noMTY = last_bonds[last_bonds["remain_nominal"]>0].reset_index()
addMTY = noMTY.copy()
addMTY["payment_type"] = "MTY"
addMTY["id_payment"] += 1
addMTY["payment_sum"] = noMTY.index.map(noMTY.remain_nominal.to_dict())
addMTY["remain_nominal"] = 0
bonds = pd.concat([bonds, addMTY], axis=0)
bonds = bonds.sort_values(by=["isin", "payment_date", "payment_type_NUM"]).drop(columns=["payment_type_NUM"]).reset_index(drop=True)
```

Мэппинг имен и кодов валют

```python
currencies = pd.read_csv("./data/Input/currencies.csv", header=0, encoding='cp1251', sep=';').set_index("Currency")["Num"]
bonds["currency"] = bonds["currency"].map(currencies).fillna(0).astype(np.int32)
```

Добавление времени выгрузки и выбор колонок

```python
bonds["export_timestamp"] = dt.now().strftime("%Y%m%d%H%M%S")
```
