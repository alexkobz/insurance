#### Шаг 3. Корректировка потока

```python
bonds = TransformedCalendar\
.merge(TransformedFintoolReferenceData[["face_value", "principal_type", "coupon_type", "issue_date"]], left_on="isin", right_index=True)\
.rename(columns={"face_value": "nominal"})
bonds["nominal"] = bonds["nominal"].fillna(0)
bonds["rate"] = bonds["rate"].fillna(0)
bonds["payment_sum"] = bonds["payment_sum"].fillna(0)
bonds['event_total_vol'] = bonds['event_total_vol'].fillna(0)
bonds["name"] = bonds["name"].str.replace('[\n\t\r;,]', '.', regex=True)
bonds["note"] = bonds["note"].str.replace('[\n\t\r;,]', '.', regex=True)
```

```python
bonds.loc[bonds["payment_type"]=="DFLT", "payment_sum"] *= -1
```

```python
bonds["payment_type_NUM"] = bonds["payment_type"].map({"CPN": 1, "MTY": 2, "PUT": 3})
```

```python
bonds["payment_sum"] = bonds["payment_sum"].where(~(bonds['event_total_vol']==0) & ~(bonds["rate"]==0), other=0)
bonds.loc[(bonds.payment_type=="CALL")&(bonds["payment_date"]>=REPORT_DATE), "payment_sum"] = 0
```

```python
call_bonds = bonds[bonds["payment_type"].isin(["CALL", "MTY"])]
cpn_bonds = bonds[~bonds["payment_type"].isin(["CALL", "MTY"])]
bonds = pd.concat([cpn_bonds, call_bonds.groupby(["isin", "payment_date"]).last().reset_index()], axis=0).sort_values(by=["isin", "payment_date", "payment_type_NUM"]).reset_index(drop=True)
```

```python
EmptyNominal = bonds[bonds.nominal==0]["isin"]
FirstMTY = bonds[(bonds["isin"].isin(EmptyNominal))&(bonds.payment_type=="MTY")].groupby("isin").first()[["rate", "payment_sum"]]
FirstMTY["nominal"] = FirstMTY["payment_sum"]/FirstMTY["rate"]
bonds.loc[bonds.nominal==0, "nominal"] = bonds[bonds.nominal==0]["isin"].map(FirstMTY.nominal.to_dict())
```

```python
bonds['id_payment'] = bonds.groupby(['isin']).cumcount() + 1
bonds["report_date"] = pd.to_datetime(REPORT_DATE)
```

```python
bonds["MTY"] = np.where(bonds.payment_type=="MTY", bonds.payment_sum, 0)
bonds["paid_nominal"] = bonds.groupby(["isin"])["MTY"].cumsum()
bonds["remain_nominal"] = bonds["nominal"] - bonds["paid_nominal"]
```

Флоатеры

```python
# floaters_isins = bonds[bonds['coupon_type'].isin(['Переменный', 'Плавающий'])].index
Floaters = bonds[bonds.index.isin(FloaterData["fintoolId"])]
Bonds = bonds[(bonds["payment_date"] > REPORT_DATE) & ~(bonds.index.isin(floaters_isins) | bonds["coupon_type"].isna())]
```

```python
Floaters_CPN_max = Floaters[Floaters.payment_type=="CPN"].groupby("isin").agg({"rate": "max"})
Floaters["CPN_max"] = Floaters["isin"].map(Floaters_CPN_max["rate"].to_dict())
```

```python
Future_CPN = Floaters[(Floaters["payment_type"]=="CPN") & (Floaters["payment_sum"]==0)].index
Floaters.loc[Future_CPN, "payment_sum"] = Floaters.loc[Future_CPN, "remain_nominal"]*Floaters.loc[Future_CPN, "CPN_max"]
```

```python
Floaters = Floaters[Floaters["payment_date"]>=REPORT_DATE]
```

#### Шаг 4. Объединение облигаций с постоянным и плавающим купоном в один датафрейм

```python
Bonds["is_floater"] = False
Floaters["is_floater"] = True
```

```python
bonds = pd.concat([Bonds, Floaters], axis=0).reset_index(drop=True)
```

Добавить отсутствующую оферту по каждому ISIN

```python
last_bonds = bonds.groupby("isin").last()
noMTY = last_bonds[last_bonds["remain_nominal"]>0].reset_index()
addMTY = noMTY.copy()
addMTY["payment_type"] = "MTY"
addMTY["id_payment"] += 1
addMTY["payment_sum"] = noMTY.index.map(noMTY.remain_nominal.to_dict())
addMTY["remain_nominal"] = 0
bonds = pd.concat([bonds, addMTY], axis=0)
bonds = bonds.sort_values(by=["isin", "payment_date", "payment_type_NUM"]).drop(columns=["payment_type_NUM"]).reset_index(drop=True)
```
